```python
#clean up the df from null data
obs_clean_pd=result_obs_pd.dropna()
obs_clean_pd.head(5)
```

```python
#created a schema / table in the postgres
#renamed the columns for consistency
obs_upload_pd = obs_clean_pd.rename(columns={"Name": "observatory_name", "Established": "year_established", "Location": "obs_location"})
obs_upload_pd.head()
```

```python
#created csv file to see what I've got
obs_upload_pd.to_csv("list_of_observatories.csv", index = False, sep=',', encoding='utf-8')
```
